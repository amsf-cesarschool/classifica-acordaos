# Fazer junção de datasets de acórdãos e lista do TRE

### Importação das libs

```python
import pandas as pd
import numpy as np

from pathlib import Path
```

## Criar dataset de todas as decisões (acórdãos e pareceres)

```python
caminho_dados = Path("../../data/raw/")
df_decisoes = pd.read_csv(caminho_dados/"tce_decisoes.csv", sep=";")
```

## Criar dataset de lista do TRE

## Juntar os datasets

```python
df_merge = df_tre.merge(df_decisoes, \
                    left_on=['COD_TCE', 'DECISAO_NUMERO', 'DECISAO_ANO'], \
                    right_on=['cod_tce', 'decisao_numero', 'decisao_ano'])
```

```python
df_merge.head()
```

```python
df_merge.info()
```

```python
df_merge[['DECISAO', 'COD_TCE', 'RESULTADO', 'cod_tce', 'peca', 'decisao_tipo', 'decisao_numero', 'decisao_ano', \
          'assunto', 'tipo_processo', 'tipo_ato', 'conselheiro', 'texto']]
```

```python
df_merge[df_merge['cod_tce']=='013381/2013']
```

### Verificar consistência dos dados de resultados

```python
df_merge['RESULTADO'].value_counts()
```

```python
df_merge.loc[df_merge['RESULTADO'] == 'IRREGUARIDADE', ['RESULTADO']] = 'IRREGULARIDADE'
```

```python
df_merge['RESULTADO'].value_counts()
```

```python
df_merge['RESULTADO'] = df_merge['RESULTADO'].apply(lambda x: x.strip())
```

```python
df_merge['RESULTADO'].value_counts()
```

```python
# Análise de resultados por tipo de ato
df_merge[['tipo_ato', 'RESULTADO', 'cod_tce']].groupby(['tipo_ato', 'RESULTADO']).count()
```

```python
# Análise de resultados por tipo de decisão
df_merge[['decisao_tipo', 'RESULTADO', 'cod_tce']].groupby(['decisao_tipo', 'RESULTADO']).count()
```

```python
# Análise de resultados por tipo de processo
df_merge[['tipo_processo', 'RESULTADO', 'cod_tce']].groupby(['tipo_processo', 'RESULTADO']).count()
```
