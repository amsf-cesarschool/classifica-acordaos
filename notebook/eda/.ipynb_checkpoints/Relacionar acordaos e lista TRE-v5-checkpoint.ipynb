### Procurar itens da lista não encontrados nas decisões

#### Itens cujos processos não foram encontrados nas decisões

```python
# Relacionados e não relacionados
df_tre_l = df_tre.merge(df_decisoes, how="left", \
                    left_on=['COD_TCE'], right_on=['cod_tce'])
```

```python
# Filtra somente os não relacionados
df_tre_l = df_tre_l[df_tre_l['cod_tce'].isna()]
df_tre_l.shape
```

##### Dos processos não encontrados, buscar por decisões
Como não foram encontrados processos que possuam acordão ou parecer para esses itens, devo fazer busca
por dados da decisão para eles.
<p>
SUSPEITA: Pode haver registro na lista com número do processo digitado errado e, talvez, sejam encontrados em outros processos
    fazendo a busca somente por número e ano da decisão

```python
# Somente relacionados por dados da decisão
df_tre_pne = df_tre_l.merge(df_decisoes, how="inner", \
                    left_on=['DECISAO_NUMERO', 'DECISAO_ANO'], \
                    right_on=['decisao_numero', 'decisao_ano'])

# Remover os que não tem número de decisão
df_tre_pne.dropna(axis=0, subset=['DECISAO_NUMERO'], inplace=True)
df_tre_pne.shape
```

Verifiquei que muito protocolos do dataframe de decisões estão com o cod_tce fora do padrão. Exemplo '14205/2012'.<br>
Fiz a correção dos registros no banco de dados através do scipt 'corrigir cod_tce de protocolos da lista do TRE.sql'<p>
Vou precisar corrigir esse dataframe também. Em futuras importações não será mais necessário, uma vez que os registros já
    encontram-se corrigidos no banco de dados

```python
# Corrigir cod_tce com formato inválido (diferente de ######/AAAA)
import os
import sys
sys.path.insert(0, os.path.abspath('../../src'))
sys.path.insert(0, os.path.abspath('../../src/preparation'))

from processa_acordao import extrai_cod_tce
#print(extrai_cod_tce(cod_tce))
def cod_tce_eh_valido(cod_tce):
    numero, _ = cod_tce.split('/')
    return len(numero) == 6

# Faz a correção
df_decisoes['cod_tce'] = df_decisoes['cod_tce'].apply(lambda x: x if cod_tce_eh_valido(x) else extrai_cod_tce(x))
```
