### Analisar e corrigir dados de decisão e resultado

```python
# Tipos de resultados e quantidade
df_tre['RESULTADO'].value_counts()
```

```python
# Remover espaços adicionais colocar em caixa alta
df_tre['RESULTADO'] = df_tre['RESULTADO'].apply(lambda x: x.strip().upper())
```

```python
# Corrigir e unificar reultado IRREGULARIDADE
df_tre.loc[df_tre['RESULTADO'].isin(['IRREGUARIDADE', 'IRREAGULARIDADE', 'IRREGULAR']), ['RESULTADO']] = 'IRREGULARIDADE'
```

```python
# Corrigir e unificar reultado REPROVAÇÃO
df_tre.loc[df_tre['RESULTADO'].isin(['PELA REPROVAÇÃO', 'RECOMENDANDO A REPROVAÇÃO']), ['RESULTADO']] = 'REPROVAÇÃO'
#df_tre.loc[df_tre['RESULTADO'] == 'RECOMENDANDO A REPROVAÇÃO', ['RESULTADO']] = 'REPROVAÇÃO'
```

```python
df_merge = df_tre.merge(df_decisoes, \
                    left_on=['COD_TCE', 'DECISAO_NUMERO', 'DECISAO_ANO'], \
                    right_on=['cod_tce', 'decisao_numero', 'decisao_ano'])
```

### Procurar itens da lista não encontrados nas decisões

#### Itens cujos processos não foram encontrados nas decisões

```python
# Relacionados e não relacionados
df_tre_l = df_tre.merge(df_decisoes, how="left", \
                    left_on=['COD_TCE'], right_on=['cod_tce'])
```

```python
# Filtra somente os não relacionados
df_tre_l = df_tre_l[df_tre_l['cod_tce'].isna()]
df_tre_l.shape
```

##### Dos processos não encontrados, buscar por decisões
Como não foram encontrados processos que possuam acordão ou parecer para esses itens, devo fazer busca
por dados da decisão para eles.
<p>
SUSPEITA: Pode haver registro na lista com número do processo digitado errado e, talvez, sejam encontrados em outros processos
    fazendo a busca somente por número e ano da decisão

```python
# Somente relacionados por dados da decisão
df_tre_pne = df_tre_l.merge(df_decisoes, how="inner", \
                    left_on=['DECISAO_NUMERO', 'DECISAO_ANO'], \
                    right_on=['decisao_numero', 'decisao_ano'])

# Remover os que não tem número de decisão
df_tre_pne.dropna(axis=0, subset=['DECISAO_NUMERO'], inplace=True)
df_tre_pne.shape
```

```python
df_tre_pne.head(1)
```

```python
df_tre_pne.sort_values(['DECISAO_ANO', 'DECISAO_NUMERO'], ascending=False)\
    [['DECISAO', 'DECISAO_NUMERO', 'DECISAO_ANO', 'PROCESSO', 'COD_TCE', \
      'cod_tce_y', 'decisao_tipo_y', 'decisao_numero_y', 'decisao_ano_y', 'peca_y']]
```

```python
# Não relacionados ordenados (decrescente) por ano de julgamento
df_tre_l[df_tre_l['ANO_DO_JULGAMENTO'] >= 2015] \
    [['ANO_DO_JULGAMENTO', 'DECISAO', 'PROCESSO', 'COD_TCE']].sort_values('ANO_DO_JULGAMENTO', ascending=False)
```

```python
# Somente relacionados por dados da decisão
df_tre_pne = df_tre_l.merge(df_decisoes, how="inner", \
                    left_on=['DECISAO_NUMERO', 'DECISAO_ANO'], \
                    right_on=['decisao_numero', 'decisao_ano'])

# Remover os que não tem número de decisão
df_tre_pne.dropna(axis=0, subset=['DECISAO_NUMERO'], inplace=True)
df_tre_pne.shape
```

```python
df_tre_pne.sort_values(['DECISAO_ANO', 'DECISAO_NUMERO'], ascending=False)\
    [['DECISAO', 'MUNICIPIO_ENTE', 'DECISAO_NUMERO', 'DECISAO_ANO', 'PROCESSO', 'COD_TCE', \
      'cod_tce_y', 'decisao_tipo_y', 'decisao_numero_y', 'decisao_ano_y', 'peca_y']]
```

Gerar arquivo com itens da planilha para os quais foram encontrados processos que possuem ato com mesmos número e ano de

### Verificar consistência dos dados de resultados
Acórdão: decisão do colegiado (plenário ou câmara) sobre julgamento de contas de gestão. Resultados possíveis
deveriam ser:
* Aprovado
* Reprovado
<p>
Parecer prévio: decisão do colegiado sobre apreciação de contas de governo (que o TCE não julga). Resultados possíveis
deveriam ser:
* Regular
* Irregular

### Buscar registros nos acórdãos e pareceres para tipo de processo CONTAS - CONTAS DE GESTÃO
Ideia é que os registros não presentes na lista do TRE tiveram contas regulares ou regulares com ressalvas

```python
# Recuperar somente os que possuem texto
df_contas = df_decisoes[(df_decisoes.tipo_processo == 'CONTAS - CONTAS DE GESTÃO') & (df_decisoes.texto.notna())]
df_contas.head()
```

```python
'REGULAR' in df_contas.loc[64, ['texto']]
```

```python
df_contas[df_contas['texto'].apply(lambda x: ' REGULAR' in x)][['cod_tce', 'peca','texto']]
```

```python
df_contas[df_contas['texto'].apply(lambda x: ' REGULARIDADE' in x)][['cod_tce', 'peca','texto']]
```

```python
# Contas regulares
df_contas[df_contas['texto'].apply(lambda x: ' REGULAR' in x and 'COM RESSALVA' not in x)][['cod_tce', 'peca','texto']]
```

```python
# Contas regulares com ressalvas
df_contas[df_contas['texto'].apply(lambda x: ' REGULAR' in x and 'COM RESSALVA' in x)][['cod_tce', 'peca','texto']]
```

```python
# Contas irregulares
df_contas[df_contas['texto'].apply(lambda x: 'IRREGULAR' in x and 'COM RESSALVA' not in x)][['cod_tce', 'peca','texto']]
```
